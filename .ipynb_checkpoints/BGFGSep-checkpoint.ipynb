```python
# -*- coding: utf-8 -*-

##################
# Standard imports
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

#############################################
# Load an example audio file with vocals.
audio_path = 'C:/Users/17205/Downloads/File/hello.mp3'
audio_data, sample_rate = librosa.load(audio_path, duration=120)

# Compute the spectrogram magnitude and phase
spec_mag, phase = librosa.magphase(librosa.stft(audio_data))

#######################################
# Display a 5-second slice of the spectrogram
time_slice = slice(*librosa.time_to_frames([0, 5], sr=sample_rate))
plt.figure(figsize=(12, 4))
librosa.display.specshow(librosa.amplitude_to_db(spec_mag[:, time_slice], ref=np.max),
                         y_axis='log', x_axis='time', sr=sample_rate)
plt.colorbar()
plt.tight_layout()

###########################################################
# The wiggly lines above are due to the vocal component.
# The goal is to separate vocals from instrumentation.
# Below steps outline the process:

# Compare frames using cosine similarity and aggregate similar frames
# by taking their (per-frequency) median value.
# To avoid being biased by local continuity, constrain similar frames to be
# separated by at least 2 seconds.
# This suppresses sparse/non-repetitive deviations from the average spectrum,
# effectively discarding vocal elements.

filtered_spec = librosa.decompose.nn_filter(spec_mag,
                                           aggregate=np.median,
                                           metric='cosine',
                                           width=int(librosa.time_to_frames(2, sr=sample_rate)))

# Ensure filter output doesn't exceed the input by taking pointwise minimum
filtered_spec = np.minimum(spec_mag, filtered_spec)

##############################################
# Use raw filter output as a mask,
# enhance results with soft-masking.

# Apply margin to reduce bleed between vocals and instrumentation masks.
# Note: margins may vary for foreground and background separation.
margin_inst, margin_voc = 2, 10
power = 2

mask_inst = librosa.util.softmask(filtered_spec,
                                  margin_inst * (spec_mag - filtered_spec),
                                  power=power)

mask_voc = librosa.util.softmask(spec_mag - filtered_spec,
                                 margin_voc * filtered_spec,
                                 power=power)

# Multiply masks with input spectrum to separate components
foreground = mask_voc * spec_mag
background = mask_inst * spec_mag

##########################################
# Plot separated foreground and background

plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
librosa.display.specshow(librosa.amplitude_to_db(spec_mag[:, time_slice], ref=np.max),
                         y_axis='log', sr=sample_rate)
plt.title('Full spectrum')
plt.colorbar()

plt.subplot(3, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(background[:, time_slice], ref=np.max),
                         y_axis='log', sr=sample_rate)
plt.title('Background')
plt.colorbar()

plt.subplot(3, 1, 3)
librosa.display.specshow(librosa.amplitude_to_db(foreground[:, time_slice], ref=np.max),
                         y_axis='log', x_axis='time', sr=sample_rate)
plt.title('Foreground')
plt.colorbar()
plt.tight_layout()
plt.savefig('Signal_Background_Foreground_Separated.jpg')
plt.show()
```
